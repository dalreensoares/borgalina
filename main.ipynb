```python
#Print CRS information
print("Small areas CRS:", small_areas.crs)
print("City line CRS:", line_2030.crs)

# Ensure both geometries are in the same CRS
if line_2030.crs != small_areas.crs:
    line_2025 = line_2030.to_crs(small_areas.crs)
else:
    line_2025 = line_2030

# Filter population data for relevant years and Capital Area
pop = pop[pop['ar'].isin([2021, 2023])].reset_index(drop=True)

# Convert smasvaedi to numeric for comparison
pop['smasvaedi'] = pd.to_numeric(pop['smasvaedi'])
pop = pop[pop['smasvaedi'] <= 2500]

# Convert back to string and ensure correct formatting
pop['smasvaedi'] = pop['smasvaedi'].astype(str).str.zfill(4)
small_areas['smsv'] = small_areas['smsv'].astype(str).str.zfill(4)

def create_population_map(year):
    """
    Create population map for total population in the selected year
    """
    # Filter population data for specific year and aggregate across all age ranges
    pop_filtered = pop[
        pop['ar'] == year  # Filter by year only
    ].groupby('smasvaedi')['fjoldi'].sum().reset_index()  # Sum across all age ranges and genders
    
    # Merge population data with small areas
    pop_smallarea = small_areas.merge(
        pop_filtered,
        left_on='smsv',
        right_on='smasvaedi',
        how='left'
    )
    
    # Create figure
    fig, ax = plt.subplots(figsize=(18, 10))
    
    # Plot the choropleth for specific NUTS3 region
    mask = pop_smallarea['nuts3'] == '001'
    pop_smallarea[mask].plot(
        column='fjoldi',
        ax=ax,
        legend=True,
        edgecolor='black',
        linewidth=0.5,
        missing_kwds={'color': 'lightgrey'},
        cmap='YlOrRd',
    )

    # Add city lines
    line_2025.plot(
        ax=ax,
        facecolor='none',
        edgecolor='black',
        linewidth=1
    )

    # Improve map appearance
    plt.title(f'Total Population Distribution (All Genders) - {year}', fontsize=14, pad=20)
    ax.axis('off')
    plt.box(False)

    # Add text showing data range
    if not pop_smallarea['fjoldi'].isna().all():
        plt.figtext(
            0.1, 0.02,
            f"Population range: {pop_smallarea['fjoldi'].min():.0f} to {pop_smallarea['fjoldi'].max():.0f}",
            fontsize=10
        )

    # Adjust the map extent to focus on the data
    pop_smallarea[mask].boundary.plot(ax=ax)
    ax.set_aspect('equal')
    plt.tight_layout()

    # Print diagnostic information
    print(f"\nYear: {year}")
    print("Number of areas with population data:",
          len(pop_smallarea[pop_smallarea['fjoldi'].notna()]))
    print("Number of areas in nuts3=='001':",
          len(pop_smallarea[mask]))
    
    plt.show()

# Create year dropdown
year_dropdown = widgets.Dropdown(
    options=[2021, 2023],
    description='Year:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='30%')
)

# Create interactive map
def update_map(year):
    plt.close('all')  # Clear previous plots
    create_population_map(year)

# Display interactive widget
interact(
    update_map,
    year=year_dropdown
)

# Function to calculate total population changes between years
def calculate_population_change():
    # Removed the kyn filter to include all gender categories
    pop_filtered = pop.groupby(['smasvaedi', 'ar'])['fjoldi'].sum().reset_index()
    
    pop_smallarea = small_areas.merge(
        pop_filtered,
        left_on='smsv',
        right_on='smasvaedi',
        how='left'
    )
    
    pop_2021 = pop_smallarea[pop_smallarea['ar'] == 2021].set_index('smsv')['fjoldi']
    pop_2023 = pop_smallarea[pop_smallarea['ar'] == 2023].set_index('smsv')['fjoldi']
    pop_change = pop_2023 - pop_2021
    
    print("\nTotal Population Changes 2021-2023 (All Genders):")
    print(f"Total change: {pop_change.sum():.0f}")
    print(f"Average change per area: {pop_change.mean():.1f}")
    print(f"Maximum increase: {pop_change.max():.0f}")
    print(f"Maximum decrease: {pop_change.min():.0f}")

# Create button for population change calculation
calculate_button = widgets.Button(
    description='Calculate Population Change',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

def on_calculate_button_clicked(b):
    calculate_population_change()

calculate_button.on_click(on_calculate_button_clicked)

# Display calculate button
display(calculate_button)
```

```python
work_places = pd.read_csv("data/fjoldi_starfandi.csv")
work_places
```

```python
#Delete irrelavent rows ie prior to 2021
work_places = work_places[work_places['ar'].isin([2021, 2022, 2023, 2024])].reset_index(drop=True)

#Delete areas outside the Capital area
work_places = work_places[work_places['smasvaedi'] <= 2500]

work_places
```

```python
# Print CRS information
print("Small areas CRS:", small_areas.crs)
print("City line CRS:", line_2030.crs)

# Ensure both geometries are in the same CRS
if line_2030.crs != small_areas.crs:
    line_2025 = line_2030.to_crs(small_areas.crs)
else:
    line_2025 = line_2030

# Filter workplace data for relevant years and Capital Area
work_places = work_places[work_places['ar'].isin([2021, 2022, 2023, 2024])].reset_index(drop=True)

# Convert smasvaedi to numeric for comparison
work_places['smasvaedi'] = pd.to_numeric(work_places['smasvaedi'])
work_places = work_places[work_places['smasvaedi'] <= 2500]

# Convert back to string and ensure correct formatting
work_places['smasvaedi'] = work_places['smasvaedi'].astype(str).str.zfill(4)
small_areas['smsv'] = small_areas['smsv'].astype(str).str.zfill(4)

# Create the main plotting function
def plot_workplace_map(year):
    plt.clf()  # Clear the current figure
    
    # Filter workplace data for specific year
    work_filtered = work_places[work_places['ar'] == year]
    
    # Merge workplace data with small areas
    work_smallarea = small_areas.merge(
        work_filtered,
        left_on='smsv',
        right_on='smasvaedi',
        how='left'
    )
    
    # Create figure
    fig, ax = plt.subplots(figsize=(20, 15))
    
    # Plot the choropleth for specific NUTS3 region
    mask = work_smallarea['nuts3'] == '001'
    work_smallarea[mask].plot(
        column='fjoldi',
        ax=ax,
        legend=True,
        edgecolor='black',
        linewidth=0.5,
        missing_kwds={'color': 'lightgrey'},
        cmap='YlOrRd',
        legend_kwds={'label': 'Number of Workplaces'}
    )

    # Add city lines
    line_2025.plot(
        ax=ax,
        facecolor='none',
        edgecolor='black',
        linewidth=1
    )

    # Improve map appearance
    plt.title(f'Workplace Distribution - {year}', fontsize=14, pad=20)
    ax.axis('off')
    plt.box(False)

    # Add text showing data range
    if not work_smallarea['fjoldi'].isna().all():
        plt.figtext(
            0.1, 0.02,
            f"Workplaces range: {work_smallarea['fjoldi'].min():.0f} to {work_smallarea['fjoldi'].max():.0f}",
            fontsize=10
        )

    # Adjust the map extent to focus on the data
    work_smallarea[mask].boundary.plot(ax=ax)
    ax.set_aspect('equal')
    plt.tight_layout()

    # Print diagnostic information
    print(f"\nYear: {year}")
    print("Number of areas with workplace data:",
          len(work_smallarea[work_smallarea['fjoldi'].notna()]))
    print("Number of areas in nuts3=='001':",
          len(work_smallarea[mask]))
    print(f"Total workplaces in region: {work_smallarea['fjoldi'].sum():.0f}")
    
    plt.show()

# Create the analysis function
def analyze_workplace_changes():
    years = sorted(work_places['ar'].unique())
    
    print("\nWorkplace Changes Between Years:")
    print("-" * 40)
    
    for i in range(len(years)-1):
        year1, year2 = years[i], years[i+1]
        
        # Merge with small areas for both years
        work_year1 = small_areas.merge(
            work_places[work_places['ar'] == year1],
            left_on='smsv',
            right_on='smasvaedi',
            how='left'
        )
        
        work_year2 = small_areas.merge(
            work_places[work_places['ar'] == year2],
            left_on='smsv',
            right_on='smasvaedi',
            how='left'
        )
        
        # Calculate changes
        total_work_year1 = work_year1['fjoldi'].sum()
        total_work_year2 = work_year2['fjoldi'].sum()
        work_change = work_year2['fjoldi'] - work_year1['fjoldi']
        
        print(f"\nChanges from {year1} to {year2}:")
        print(f"Total workplaces {year1}: {total_work_year1:,.0f}")
        print(f"Total workplaces {year2}: {total_work_year2:,.0f}")
        print(f"Net change: {total_work_year2 - total_work_year1:,.0f}")
        print(f"Percentage change: {((total_work_year2 - total_work_year1) / total_work_year1 * 100):.1f}%")
        print(f"Average change per area: {work_change.mean():.1f}")
        print(f"Maximum increase in an area: {work_change.max():.0f}")
        print(f"Maximum decrease in an area: {work_change.min():.0f}")

# Create and display widgets
year_selector = widgets.Dropdown(
    options=sorted(work_places['ar'].unique()),
    description='Select Year:',
    style={'description_width': 'initial'}
)

analyze_button = widgets.Button(
    description='Analyze Changes',
    style={'description_width': 'initial'}
)

# Define widget callbacks
def on_year_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        plot_workplace_map(change['new'])

def on_analyze_click(b):
    analyze_workplace_changes()

# Connect callbacks
year_selector.observe(on_year_change, names='value')
analyze_button.on_click(on_analyze_click)

# Display widgets and initial map
display(widgets.VBox([year_selector, analyze_button]))
plot_workplace_map(year_selector.value)
```

Based on the population distribution map and population change data, we can make several observations about the bus system implications:

1. The darkest red area in the center appears to be the most densely populated region, suggesting this area likely needs the highest frequency of bus service and greatest number of stops to serve the large population base.

2. The orange-colored areas around the periphery show moderate population density, indicating a need for reliable transit connections between these residential zones and the central area.

3. With a total population increase of 9,690 between 2021-2023, there's likely growing pressure on the existing bus network. Some key implications:
   - The maximum increase of 1,225 people in a single area suggests at least one zone needs significant review of its transit capacity
   - Even areas with moderate growth (average +73.4 people per area) may need service adjustments
   - The relatively small maximum decrease (-122) suggests no areas need major service reductions

4. The scattered lighter colored areas with lower population density would likely benefit from:
   - Strategic bus stop placement at community hubs
   - Less frequent but reliable scheduled service
   - Possible feeder routes connecting to main transit corridors

5. The stark contrast between the central red zone and surrounding areas suggests a need for a hub-and-spoke bus network design, with frequent service in the core and strategic connector routes to the peripheral regions.

Conclusion on Bus Stop Distribution:
The adequacy of bus stop distribution (represented by black circles) can be evaluated in relation to population density and workplace changes:

1. **Population Density:**
   - Darker areas on the population map indicate higher density. These areas appear to have a higher concentration of bus stops, which aligns with the need for greater public transport accessibility in densely populated areas.
   - However, some outlying areas with lighter population density have fewer bus stops, which may result in longer travel times or less accessibility for residents.

2. **Workplace Distribution:**
   - The growth in workplaces from 2021 to 2024 shows significant changes, particularly in areas experiencing the largest increases (e.g., a maximum of 444 workplaces from 2021 to 2022).
   - If bus stops are not located near areas with the largest workplace increases, employees may face challenges commuting efficiently.

3. **Geographic Spread:**
   - The bus stops appear well-distributed in the central urban areas but seem sparse in the outer areas, despite some having moderate or significant workplace and population growth.

4. **Accessibility:**
   - While the maps suggest a correlation between higher density (population/workplace) and bus stop locations, the adequacy of coverage depends on whether residents and workers in outlying regions are within a reasonable walking distance to a stop.
